# Numpy Implementation

```python
import numpy as np

# Create a 2D Numpy array of size 1x3 with elements of your choice
arr1 = np.array([[1, 2, 3]])

# Create a Numpy array of length 50 with zeroes as its elements
arr2 = np.zeros(50)

# Create a Numpy array of length 3x2 with elements of your choice
arr3 = np.array([[1, 2],
                 [3, 4],
                 [5, 6]])

# Multiply arr1 and arr3 using Numpy functions
arr4 = np.dot(arr1, arr3)

# Change 5th element of arr2 to a different number
arr2[4] = 5

# The test condition is already provided, so we'll leave it as is
if np.shape(arr4)==(1,2) and arr2[4]!=0:
  print("Passed")
else:
  print("Fail")
```

```python
import numpy as np

# Define the 3x3 identity matrix
I = np.eye(3)

# Compute 9I + 1
result_matrix = 9 * I + 1

# Perform dot product of I and (9I + 1)
dot_product = np.dot(I, result_matrix)

print("Dot product of I and 9I + 1:")
print(dot_product)
```

# Pandas Implementation

```python
import pandas as pd

# Create a DataFrame from the dictionary
data = {
    'Name': ['Ramesh', 'Mahesh', 'Suresh'],
    'Age': [25, None, 35],
    'City': ['Bangalore', 'Mumbai', 'Delhi']
}

df = pd.DataFrame(data)

# Fill missing values in the 'Age' column with the mean age
mean_age = df['Age'].mean()
df['Age'].fillna(mean_age, inplace=True)

print("DataFrame after filling missing values:")
print(df)
```

# Matplotlib Implementation

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 3, 4])
ypoints = np.array([2, 4, 6, 8])

# Plot these points without drawing a line
plt.scatter(xpoints, ypoints, color='black')  # Using scatter plot without a line

plt.title('Scatter Plot without Line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# Plotting with marker: Plot these points with a star marker
plt.scatter(xpoints, ypoints, marker='*')

plt.title('Scatter Plot with Star Marker')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# Using fmt format, add circular marker, red color, and dashed line
plt.plot(xpoints, ypoints, 'ro--')

plt.title('Plot with Circular Marker, Red Color, and Dashed Line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# Create a scatter plot for xpoints and ypoints
plt.scatter(xpoints, ypoints)

plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# Set color to the scatter plot. Blue, Green, Red, and Yellow color for each point respectively
colors = ['blue', 'green', 'red', 'yellow']
plt.scatter(xpoints, ypoints, color=colors)

plt.title('Scatter Plot with Different Colors')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# Miscellaneous Modules Implementation

```python
import random
import numpy as np

# Set the seed of random to 20
random.seed(20)

arr1 = np.array([1, 24, 31, 45, 73, 81, 94, 25])

# Using the random module, pick 4 different random numbers from arr1 and return their sum
random_indices = random.sample(range(len(arr1)), 4)  # Pick 4 unique indices
selected_numbers = arr1[random_indices]
sum_selected = np.sum(selected_numbers)

print("Randomly picked numbers:", selected_numbers)
print("Sum of the randomly picked numbers:", sum_selected)
```
